## Изменение файла settings.json

```python
import json

settings_path = "/root/.config/Ultralytics/settings.json"

with open(settings_path, "r") as f:
    settings = json.load(f)

settings["datasets_dir"] = "/kaggle/input/logo-detection"

with open(settings_path, "w") as f:
    json.dump(settings, f, indent=4)

print("settings.json обновлен!")
```

```python
import yaml

data_yaml_path = "/kaggle/input/logo-detection/done/data.yaml"

with open(data_yaml_path, "r") as f:
    data_config = yaml.safe_load(f)

data_config["path"] = "/kaggle/input/logo-detection/done"

new_yaml_path = "/kaggle/working/data.yaml"
with open(new_yaml_path, "w") as f:
    yaml.dump(data_config, f, default_flow_style=False)

print("Файл data.yaml обновлен и сохранен в /kaggle/working/data.yaml")
```
